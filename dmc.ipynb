```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
```

```python
img = cv2.cvtColor(cv2.imread('/KA.HA3.31.tiff'), cv2.COLOR_BGR2RGB)
```

```python
print(img.shape)
```

```python
image_matrix = np.array(img)
```

```python
pca_components = 15
```

```python
pca = PCA(n_components=pca_components)
reduced_image = pca.fit_transform(image_matrix.reshape(-1, 3))
```

```python
reconstructed_image = pca.inverse_transform(reduced_image).reshape(img.shape)
```

```python
plt.imshow(reconstructed_image.astype(np.uint8))
```
